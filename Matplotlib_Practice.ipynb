```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y =  x**2
```

```python
# Functional Method
plt.plot(x,y)
plt.xlabel('x-data')
plt.ylabel('y-data')
plt.title('Functional Method')
plt.show()
```

```python
# multiple plots
plt1 = plt.subplot(1,2,1)
plt1.plot(x,y)
plt2 = plt.subplot(1,2,2)
plt2.plot(y,x)
plt.show()
```

```python
## OO (Object Oriented Method)

fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.9,0.9])
axes1.set_xlabel('AX1')
axes1.set_ylabel('AY1')
axes1.set_title('AXES1')
axes1.plot(x,y)

axes2 = fig.add_axes([0.2,0.4,0.8,0.4])
axes2.set_xlabel('AX2')
axes2.set_ylabel('AY2')
axes2.set_title('AXES2')
axes2.plot(x,y)

plt.tight_layout()
```

```python
## OO (Object Oriented Method) - Tuple Unpacking
fig,axes = plt.subplots(nrows = 1, ncols = 2,figsize = (10,10))

axes[0].plot(x,y,label = '1')
axes[1].plot(x,y,label = '2')
axes[1].set_xlim([0,3])
axes[1].set_ylim([0,15])

axes[0].legend()
axes[1].legend(loc = 2)
```
